# Number of records by reason for producing embroys or storing eggs, 1991-2018

```python
#Reading the first table from the excel
sheet_1 = pd.read_excel("family-formations-2018-complete-underlying-data-tables.xlsx", sheet_name=3,skiprows=3)
```

```python
#data cleaning
sheet_1 = sheet_1[:-3]
for col in sheet_1.columns:
    sheet_1[col] = sheet_1[col].fillna(0)
sheet_1 = sheet_1.replace('<5',5)
```

```python
sheet_1
```

```python
#fresh embroys versus IVF frozen embroys over the years. It implies that IVF-fresh embryos
#are increasing and decreasing in cyclic pattern, and the number of IVF frozen embryos are increasing consitently and growing exponentially,
#may be because the success rate of the IVF - frozen embryos over the years is more compare to IVF fresh embryos.
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='IVF - fresh embryos', ax=ax)
sheet_1.plot(x='Year', y='IVF - frozen embryos', ax=ax, secondary_y=True)
```

```python
#ICSI fresh embroys were increasing consistently over the years till 2014 then it started to decline from then on till now. 
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='ICSI - fresh embryos', ax=ax)
```

```python
#DI stimulated versus unstimulated. as we can understand over the years Donar Insimination stimulated and unstimulated
#are steadily decreasing till 2005 from 2005 the numbers are increasing over the recent years. 
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='DI - stimuated', ax=ax)
sheet_1.plot(x='Year', y='DI - unstimulated', ax=ax)
```

```python
#donations over the years increasing and egg sharing was increasing till 2004 then it started to declining continously 
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='Donation', ax=ax)
sheet_1.plot(x='Year', y='Egg share', ax=ax, secondary_y=True)
```

```python
#egg freeze cycles were growing exponentially over the years and egg thaw cycles were constant till 2009 and 
#the numbers of egg thaw cycles from 2009 growing exponentially.
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='Egg freeze cycles', ax=ax)
sheet_1.plot(x='Year', y='Egg thaw cycles', ax=ax, secondary_y=True)
```

```python
#Embryo storage numbers are growing exponentially over the years and egg and embryo storage are consistantly less than five
#per year over the years
fig, ax = plt.subplots()
sheet_1.plot(x='Year', y='Embryo storage', ax=ax)
sheet_1.plot(x='Year', y='Egg and embryo storage', ax=ax, secondary_y=True)
```

As we understand by looking at the above variables in the first table over the years the number of IVF frozen embryos,
Donations, Egg freeze cycles and embryo storages are increasing consistently and growing in exponential pattern that implies frozen embroys in In vitro fertilization(IVF) are successful in giving birth. and also the egg freezing cycles are also increasing. as we can understand by this that in the coming years the storages of embroys and IVF frozen embroys plays a very important role and also using this statistics we can eliminate or automate the certain steps in birth cycle using advanced technology for high precision.   

# Treatment cycles by partner type 2000-2018

```python
#Reading the second table from the excel
sheet_2 = pd.read_excel("family-formations-2018-complete-underlying-data-tables.xlsx", sheet_name=4,usecols=None, skiprows=3)
```

```python
#data cleaning
sheet_2.rename(columns={'IVF': 'IVF_Male', 'Unnamed: 2': 'IVF_Female', 'Unnamed: 3': 'IVF_None', 'Unnamed: 4':'IVF_surrogate','Unnamed: 5':'IVF_Not_recorded'}, inplace=True)
sheet_2.rename(columns={'DI': 'DI_Male', 'Unnamed: 7': 'DI_Female', 'Unnamed: 8': 'DI_None', 'Unnamed: 9':'DI_Not_recorded'}, inplace=True)
sheet_2 = sheet_2.drop([0])
for col in sheet_2.columns:
    sheet_2[col] = sheet_2[col].fillna(0)
sheet_2 = sheet_2.replace('<5',5)
```

```python
sheet_2
```

```python
# IVF_Male partner versus IVF_Female partner, the cases in both criteria are growing and in the IVF_Male 
#cases numbers are high linearly and in IVF_Female number of cases are low but growing consistently.
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='IVF_Male', ax=ax)
sheet_2.plot(x='Year', y='IVF_Female', ax=ax, secondary_y=True)
```

```python
#IVF_No partner versus IVF_Surrogate. the number of cases of IVF with no partner is increasing over the years.
#and also the IVF with surrogate cases are increasing in less numbers there is some fluctuations in the number of cases in 
# 2002, 2012 and 2016.
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='IVF_None', ax=ax)
sheet_2.plot(x='Year', y='IVF_surrogate', ax=ax, secondary_y=True)
```

```python
# DI_male partner versus DI_Female partner. the number of cases of DI_Male are decreasing rapidly over the years,
#the numbers are almost disappearing. and in the case of DI_Female(on right), we can see that there is a steady 
#growth in the number of cases. 
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='DI_Male', ax=ax)
sheet_2.plot(x='Year', y='DI_Female', ax=ax, secondary_y=True)
```

```python
#DI with no partner has strong deviation over the years but the numbers from 2000-2018 are similar
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='DI_None', ax=ax)
```

```python
# IVF_Male versus DI Male, as we can see  the numbers for DI Male is decreasing continously over the years
#from 2000-2018. and in IVF_Male case, the numbers are increasing linearly.
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='IVF_Male', ax=ax)
sheet_2.plot(x='Year', y='DI_Male', ax=ax, secondary_y=True)
```

```python
# IVF_Female versus DI Female Versus, as we can see in this graph that both IVF_Female and DI_Female are growing over the years
fig, ax = plt.subplots()
sheet_2.plot(x='Year', y='IVF_Female', ax=ax)
sheet_2.plot(x='Year', y='DI_Female', ax=ax, secondary_y=True)
```

In this table we understand that in In-Vitro-Fertilizatio(IVF) in both male and female partners, the numbers are increasing consistently, number of cases of IVF_Male are more and in IVF_Female numbers are less but increasing consistently. and IVF with no partner are also increasing slowly and in case of IVF_surrogate the numbers are fluctuating a bit but over the years it increasing slowly. In Donor Insimination(DI) in male partner are decreasing rapidly over the years and in female partner cases are increasing slowly.

# Patient age by partner type with break down by age bands, 2018

```python
sheet_3 = pd.read_excel("family-formations-2018-complete-underlying-data-tables.xlsx", sheet_name=5,usecols=None, skiprows=3)
```

```python
sheet_3
```

```python
#data cleaning
sheet_3.rename(columns={'IVF': 'IVF_average patient age', 'Unnamed: 2': 'IVF_Standard deviation of patient age', 'Unnamed: 3': 'IVF_under 35', 'Unnamed: 4':'IVF_35-37','Unnamed: 5':'IVF_38-39','Unnamed: 6':'IVF_40-42','Unnamed: 7':'IVF_43-44','Unnamed: 8':'IVF_Over 44'}, inplace=True)
sheet_3.rename(columns={'DI': 'DI_average patient age', 'Unnamed: 10': 'DI_Standard deviation of patient age', 'Unnamed: 11': 'DI_under 35', 'Unnamed: 12':'DI_35-37','Unnamed: 13':'DI_38-39','Unnamed: 14':'DI_40-42','Unnamed: 15':'DI_43-44','Unnamed: 16':'DI_Over 44'}, inplace=True)
sheet_3 = sheet_3.drop([0])
for col in sheet_3.columns:
    sheet_3[col] = sheet_3[col].fillna(0)
```

```python
sheet_3
```

```python
ax = sheet_3[sheet_3.columns].plot(kind='bar', title ="Patient age", figsize=(15, 10), legend=True, fontsize=16)
ax.set_xlabel("Partner", fontsize=12)
ax.set_ylabel("Numbers", fontsize=12)
plt.show()
```

As we saw the patient age distribution in the above table the ideal age patients choose for IVF for a male partner should be 35 with deviation of +/- 4, for a female partner  34 with deviation of +/- 4, with no partner it would be 39 with deviation of +/- 4, for surrogate patients it would be 35 with deviation of +/- 6. the average age for any kind of partner for IVF is 35 with deviation of 4. and for Donor Insimination the ideal age patients choose for male partner is 34 with deviation of +/- 4, for female partner 33 with deviation of +/- 4, for no partner is 37 with deviation of +/- 4. the average age for Donor Insimination for any partner is 34 with deviation of +/- 4.
